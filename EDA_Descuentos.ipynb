```python
import pandas as pd
import numpy as np
```

```python
df_maestro_articulos = pd.read_excel( 'maestro.xlsx' , sheet_name='Sheet1',dtype={'Ean':str}).fillna('')
```

```python
df_products_price = pd.read_excel( 'Productos.xlsx' , sheet_name='Agrup.',dtype={'Ean':str},decimal='.').fillna('')
df_products_desc = pd.read_excel( 'Productos.xlsx' , sheet_name='Des Max',dtype={'Ean':str},decimal='.').fillna('')

df_products_price['Precio'] = df_products_price['Precio'].apply(lambda x: float(x))

df_products_price = df_products_price.rename(columns={'Precio':'Price KG'})
df_products_price['Price LB'] = df_products_price['Price KG']/2.205
```

Comprueba que los precios de las familias sean iguales

1.   Se valida los diferentes precios de los productos entre familias
2.   Se obtiene el dataframe con los productos correctos
3. Los productos con precios diferentes se guardan en un dataframe aparte


```python
# 1 -> Se hace el recorte para comparar solo los productos que estan en el excel
df_to_compare_ma = df_maestro_articulos.loc[df_maestro_articulos['Ean'].isin(df_products_price['Ean'])]
df_vs = pd.merge(df_to_compare_ma,df_products_price, on='Ean',how='left')
df_vs = df_vs.sort_values(by='Agrup.')
```

```python
df_price_fam = df_vs[['Agrup.','Price KG_x']].groupby('Agrup.')['Price KG_x'].nunique().to_frame()
fam_diferent_price_one = list(df_price_fam[df_price_fam['Price KG_x']==2].index.values)
fam_diferent_price_more = list(df_price_fam[df_price_fam['Price KG_x']>2].index.values)
```

```python
list_product_incorrectos = []
for agrupacion in fam_diferent_price_one:
  cod_incorrecto = df_vs[df_vs['Agrup.']== agrupacion]
  cod_incorrecto = cod_incorrecto[['Ean','Price KG_x']]
  # Toma el valor mayor en caso de que el codigo incorrecto se mayor a los otros, sino toma el menor
  indice_valor_diferente = (cod_incorrecto['Price KG_x'] != cod_incorrecto['Price KG_x'].iloc[0]).idxmax()
  indice_valor_diferente #max 8 min 15
  valor_diferente = cod_incorrecto.loc[indice_valor_diferente]['Price KG_x']
  if len(cod_incorrecto[cod_incorrecto['Price KG_x']==valor_diferente])==1:
    cod_incorrecto_x = cod_incorrecto[cod_incorrecto['Price KG_x']==valor_diferente].Ean
    list_product_incorrectos.append(cod_incorrecto_x.values[0])
    print(indice_valor_diferente)
  else:
    indice_valor_diferente = (cod_incorrecto['Price KG_x'] != cod_incorrecto['Price KG_x'].iloc[0]).idxmin()
    indice_valor_diferente #max 8 min 15
    valor_diferente = cod_incorrecto.loc[indice_valor_diferente]['Price KG_x']
    if len(cod_incorrecto[cod_incorrecto['Price KG_x']==valor_diferente])==1:
      cod_incorrecto_x = cod_incorrecto[cod_incorrecto['Price KG_x']==valor_diferente].Ean
      list_product_incorrectos.append(cod_incorrecto_x.values[0])
      print(indice_valor_diferente)


list_product_incorrectos
# cod_incorrecto
# indice_valor_diferente
fam_diferent_price_one
```

```python
ean_cls = df_vs.loc[df_vs['Agrup.'].isin(fam_diferent_price_one + fam_diferent_price_more)] # Se quitan estos codigos
df_products_price_cls = df_vs[['Agrup.','CodSap','Ean','Descripción_y','Um','Linea','Price KG_x']]
df_products_price_cls['Price LB'] = df_products_price_cls['Price KG_x']/2.205
df_products_price_cls=df_products_price_cls.rename(columns={'Price KG_x':'Price KG','Descripción_y':'Descripción'})
df_products_price_cls
```

```python
df_maestro_articulos['Agrup.'] = 'INDIVIDUAL'
df_maestro_articulos = df_maestro_articulos.loc[~df_maestro_articulos.Ean.isin(df_products_price_cls.Ean)]
ean_cls
```

Unificar los dataframes


1.   Se le agrega la columna Agrup. al dataframe maestro
2.   Se quita los elementos que estan en el excel al dataframe maestro para que no se repitan
3. Se los concatenan
4. Se mueve la columna Agrup al inicio




```python
df_refresh =  pd.concat([df_maestro_articulos,df_products_price_cls])
df_refresh = df_refresh.loc[~df_refresh['Ean'].isin(ean_cls.Ean)]
column_agrup = df_refresh.pop('Agrup.')
df_refresh.insert(0,'Agrup.',column_agrup)
df_refresh
```
